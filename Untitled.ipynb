```python
from pydfs_lineup_optimizer import get_optimizer, Site, Sport
```

```python
optimizer = get_optimizer(Site.FANDUEL, Sport.BASEBALL)
```

```python
optimizer
```

```python
optimizer.load_players_from_csv("C:\\Users\\wallg\\MLBDFSLineupOptimizer\\Untitled spreadsheet - Sheet1 (1).csv")
```

```python
lineups = optimizer.optimize(n=150, max_exposure=0.3)
```

```python
lineups
```

```python
for lineup in optimizer.optimize(n=150, max_exposure=0.3):
    print(lineup)
    print(lineup.players)  # List of players
    print(lineup.fantasy_points_projection)
    print(lineup.salary_costs)
```
